```python
experiments = list()


for i in range(1, 10**7):
    exp = list()
    # 1 - теорвер, 0 - другие задачи
    x = [1 for i in range(12)] + [0 for i in range(8)]
    random.shuffle(x)
    max_n = len(x) - 1
    for i in range(4):
        exp.append(x.pop(random.randint(0, max_n)))
        max_n -= 1
    experiments.append(exp)
    
exp_stats = dict()

for e in experiments:
    p_outcome = sum(e)
    if p_outcome in exp_stats.keys():
        exp_stats[p_outcome] += 1
    else:
        exp_stats[p_outcome] = 1
```

```python
ans = exp_stats[1] / sum(exp_stats.values())
print(ans)
print(round(ans,2))
```

```python
total_exp = list()

for i in range(10**7):
    exp = list()
    mag = ['p'] * 5 + ['a'] * 5 + ['g'] * 5
    random.shuffle(mag)
    max_n = len(mag) - 1
    for i in range(5):
        exp.append(mag[random.randint(0, max_n)])
    total_exp.append(exp)
```

```python
exp_stats = dict()
flowers = {'p','a','g'}

for exp in total_exp:
    if flowers == set(exp):
        if '3_types' in exp_stats.keys():
            exp_stats['3_types'] += 1
        else:
            exp_stats['3_types'] = 1
    else:
        if 'non_3_types' in exp_stats.keys():
            exp_stats['non_3_types'] += 1
        else:
            exp_stats['non_3_types'] = 1
```
