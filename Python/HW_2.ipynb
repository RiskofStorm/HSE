```python
from datetime import datetime

def newspapers_datetime_print(datetime: datetime, zpn='#') -> None:
    
    print(f'The Moscow Times - {datetime.strftime(f"%A, %B %{zpn}d, %Y")}')
    print(f'The Guardian - {datetime.strftime(f"%A, %d.%m.%y")}')
    print(f'Daily News - {datetime.strftime(f"%A, %d %B %Y")}')
```

```python
newspapers_datetime_print(datetime.now())
```

```python
def date_range(start_date: datetime, end_date: datetime) -> list:
    if start_date > end_date:
        return list()
    
    start = [start_date.year, start_date.month, start_date.day]
    end = [end_date.year, end_date.month, end_date.day + 1]
    
    ret_val = list()
    
    while start != end:
        try:
            ret_val.append(datetime(start[0], start[1], start[2]).strftime('%Y-%m-%d'))
            start[2] += 1
        except ValueError as e:
            # ->> next month
            if str(e).startswith('day') and start[1] < 12:
                start[1], start[2] = start[1] + 1, 1
            # ->> next year
            elif str(e).startswith('day') and start[1] == 12:
                start[0], start[1], start[2] = start[0] + 1, 1, 1
    return ret_val
```

```python
start_date = datetime(2022,1,1)
end_date = datetime(2022,1,3)

date_range(start_date, end_date)
```

```python
import pprint as pp
start_date = datetime(2022,1,1)
end_date = datetime(2024,1,3)

pp.pprint(date_range(start_date, end_date), compact=1)
```
